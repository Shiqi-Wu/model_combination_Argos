```python
import numpy as np
import matplotlib.pyplot as plt

# 设置参数
a = 1
b = 1
omega_1 = 2
omega_2 = 3

# 定义x的范围
x = np.linspace(-6*np.pi, 6*np.pi, 400)

# 计算y值
y = a * np.sin(omega_1 * x) + b * np.cos(omega_2 * x)

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$a\sin(\omega_1 x) + b\cos(\omega_2 x)$')
plt.title('$a\sin(\omega_1 x) + b\cos(\omega_2 x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
